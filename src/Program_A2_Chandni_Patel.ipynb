```python
class image_processor:
    
    #initialling the image processor
    def __init__(self, img, img2):
        self.TITLE = 'Corner Detection 1'
        self.NoiseBar = "Variance"
        self.SmoothingBar = "Smoothing"
        self.KBar = "k"
        self.WinSizeBar = "WinSize"
        self.ThresholdBar = "Threshold"
        
        self.TITLE2 = 'Corner Detection 2'        
        self.NoiseBar2 = "Variance"
        self.SmoothingBar2 = "Smoothing"
        self.KBar2 = "k"
        self.WinSizeBar2 = "WinSize"
        self.ThresholdBar2 = "Threshold"
        
        self.support_keys = {ord('h'),ord('i'),ord('w'),ord('1'),ord('2'),ord('3')}
        self.img_h()
        
        self.in_img = cv2.imread(img)    
        self.gray_img = cv2.cvtColor(self.in_img, cv2.COLOR_BGR2GRAY)
        self.curr_img = self.in_img
        
        self.in_img2 = cv2.imread(img2)    
        self.gray_img2 = cv2.cvtColor(self.in_img2, cv2.COLOR_BGR2GRAY)
        self.curr_img2 = self.in_img2
        
        self.get_preview()   

    #showing the image and waiting for key press
    def get_preview(self):           
        cv2.imshow(self.TITLE, self.curr_img) 
        cv2.createTrackbar(self.NoiseBar, self.TITLE, 0, 10, self.tracker) 
        cv2.createTrackbar(self.SmoothingBar, self.TITLE, 0, 100, self.tracker) 
        cv2.createTrackbar(self.KBar, self.TITLE, 4, 50, self.tracker) 
        cv2.createTrackbar(self.WinSizeBar, self.TITLE, 9, 100, self.tracker) 
        cv2.createTrackbar(self.ThresholdBar, self.TITLE, 500, 1000, self.tracker) 
        
        cv2.imshow(self.TITLE2, self.curr_img2) 
        cv2.createTrackbar(self.NoiseBar2, self.TITLE2, 0, 10, self.tracker) 
        cv2.createTrackbar(self.SmoothingBar2, self.TITLE2, 0, 100, self.tracker) 
        cv2.createTrackbar(self.KBar2, self.TITLE2, 4, 50, self.tracker) 
        cv2.createTrackbar(self.WinSizeBar2, self.TITLE2, 9, 100, self.tracker) 
        cv2.createTrackbar(self.ThresholdBar2, self.TITLE2, 500, 1000, self.tracker) 
        
        while True:   
            k = cv2.waitKey(10) & 0xFF
            if k == 27:
                self.exit()
                break                
            if k in self.support_keys:
                self.command_key(k)               

    #controlling all support keys
    def command_key(self, k):
        k_const = 100
        threshold_const = 100000000 
        
        if k == ord('h'):
            self.img_h()     
        if k == ord('i'):
            self.img_i() 
        if k == ord('w'):
            self.img_w()
            
        if k == ord('1'):
            variance = cv2.getTrackbarPos(self.NoiseBar, self.TITLE)
            smoothing = cv2.getTrackbarPos(self.SmoothingBar, self.TITLE)
            k = cv2.getTrackbarPos(self.KBar, self.TITLE)/k_const
            windowsize = cv2.getTrackbarPos(self.WinSizeBar, self.TITLE)        
            threshold = cv2.getTrackbarPos(self.ThresholdBar, self.TITLE)*threshold_const   
            self.curr_img, finalCorners, features = self.detectCorners(self.gray_img, variance, smoothing, k, windowsize, threshold)
            cv2.imshow(self.TITLE, self.curr_img)  
            
            variance = cv2.getTrackbarPos(self.NoiseBar2, self.TITLE2)
            smoothing = cv2.getTrackbarPos(self.SmoothingBar2, self.TITLE2)
            k = cv2.getTrackbarPos(self.KBar2, self.TITLE2)/k_const
            windowsize = cv2.getTrackbarPos(self.WinSizeBar2, self.TITLE2)        
            threshold = cv2.getTrackbarPos(self.ThresholdBar2, self.TITLE2)*threshold_const   
            self.curr_img2, finalCorners2, features2 = self.detectCorners(self.gray_img2, variance, smoothing, k, windowsize, threshold)
            cv2.imshow(self.TITLE2, self.curr_img2)  
            
        if k == ord('2'):
            variance = cv2.getTrackbarPos(self.NoiseBar, self.TITLE)
            smoothing = cv2.getTrackbarPos(self.SmoothingBar, self.TITLE)
            k = cv2.getTrackbarPos(self.KBar, self.TITLE)/k_const
            windowsize = cv2.getTrackbarPos(self.WinSizeBar, self.TITLE)        
            threshold = cv2.getTrackbarPos(self.ThresholdBar, self.TITLE)*threshold_const   
            plot_img1, finalCorners1, features1 = self.detectCorners(self.gray_img, variance, smoothing, k, windowsize, threshold)

            variance = cv2.getTrackbarPos(self.NoiseBar2, self.TITLE2)
            smoothing = cv2.getTrackbarPos(self.SmoothingBar2, self.TITLE2)
            k = cv2.getTrackbarPos(self.KBar2, self.TITLE2)/k_const
            windowsize = cv2.getTrackbarPos(self.WinSizeBar2, self.TITLE2)        
            threshold = cv2.getTrackbarPos(self.ThresholdBar2, self.TITLE2)*threshold_const   
            plot_img2, finalCorners2, features2 = self.detectCorners(self.gray_img2, variance, smoothing, k, windowsize, threshold)

            self.compareFeatures(plot_img1, finalCorners1, features1, plot_img2, finalCorners2, features2)
            
        if k == ord('3'):
            variance = cv2.getTrackbarPos(self.NoiseBar, self.TITLE)
            smoothing = cv2.getTrackbarPos(self.SmoothingBar, self.TITLE)
            k = cv2.getTrackbarPos(self.KBar, self.TITLE)/k_const
            windowsize = cv2.getTrackbarPos(self.WinSizeBar, self.TITLE)        
            self.curr_img = self.detectCorners_OpenCV(self.gray_img, variance, smoothing, k, windowsize)
            cv2.imshow(self.TITLE, self.curr_img)  
            
            variance = cv2.getTrackbarPos(self.NoiseBar2, self.TITLE2)
            smoothing = cv2.getTrackbarPos(self.SmoothingBar2, self.TITLE2)
            k = cv2.getTrackbarPos(self.KBar2, self.TITLE2)/k_const
            windowsize = cv2.getTrackbarPos(self.WinSizeBar2, self.TITLE2)        
            self.curr_img2 = self.detectCorners_OpenCV(self.gray_img2, variance, smoothing, k, windowsize)
            cv2.imshow(self.TITLE2, self.curr_img2)  
            

    #exit
    def exit(self):
        cv2.destroyAllWindows()

    #tracker event not using
    def tracker(self, a):
        return
    
    #display a short description of the program, command line argument and support support_keys
    def img_h(self):
        print("\nSupport support_keys are as below:")         
        print("1 -> Apply Harris Corner Detection on both images")
        print("2 -> Compare Features of both images")
        print("3 -> OpenCV corner detection for verification")
        print("h -> Display help")
        print("i -> Reload the original images") 
        print("w -> Save the current images")
        print("ESC -> Exit\n\n")
        
    #reload the original images
    def img_i(self):
        self.curr_img = np.copy(self.in_img)
        cv2.imshow(self.TITLE, self.curr_img)         
        self.curr_img2 = np.copy(self.in_img2)
        cv2.imshow(self.TITLE2, self.curr_img2) 

    #save the current processed images
    def img_w(self):
        cv2.imwrite('out1.jpg', self.curr_img)        
        cv2.imwrite('out2.jpg', self.curr_img2)
        
    #get image gradient x and y
    def getGradient(self, img):        
        k_x, k_y = cv2.getDerivKernels(0, 1, 3, True)
        dx = cv2.filter2D(img, -1, k_y)
        dx = cv2.filter2D(dx, -1, k_x.T)
        dx = np.float64(dx) * 255.0 / dx.max()
        k_x, k_y = cv2.getDerivKernels(1, 0, 3, True)
        dy = cv2.filter2D(img, -1, k_x.T)
        dy = cv2.filter2D(dy, -1, k_y)
        dy = np.float64(dy) * 255.0 / dy.max()   
        return dx, dy
        
    #add gaussian noise
    def addNoise(self, img, mean, variance):
        if variance == 0:
            return img
        r, c= img.shape
        gauss = np.random.normal(mean, variance**0.05, (r, c))
        img = img + gauss.reshape(r, c)
        return img.astype(np.uint8)

    #apply gaussian smoothing
    def applySmoothing(self, img, smoothing):
        if smoothing == 0:
            return img
        if smoothing % 2 == 0:
            smoothing+=1
        img = cv2.GaussianBlur(img, (smoothing,smoothing), (smoothing / 5), (smoothing / 5))
        return img        
        
    #detect corner using OpenCV
    def detectCorners_OpenCV(self, img, variance, smoothing, k, windowsize):
        img = self.addNoise(img, 0, variance)
        img = self.applySmoothing(img, smoothing)
        plot_img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)  
        if windowsize % 2 == 0:
            windowsize += 1
        dst = cv2.cornerHarris(img,windowsize,3,k)
        plot_img[dst>0.01*dst.max()]=[0,0,255]
        return plot_img
        
        
    #Detect corners and plot them
    def detectCorners(self, img, variance, smoothing, k, windowsize, threshold): 
        print("Plotting...")
        img = self.addNoise(img, 0, variance)
        img = self.applySmoothing(img, smoothing)
        
        plot_img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)        
        height = img.shape[0]
        width = img.shape[1]  
        features = []
        
        #Calculate window offset
        if windowsize % 2 == 0:
            windowsize += 1
        offset = windowsize//2
              
        #Get gradient
        dx, dy = self.getGradient(img)
        Ixx = dx*dx
        Iyy = dy*dy
        Ixy = dx*dy        
        
        cornerList = self.harrisCorners(height, width, Ixx, Iyy, Ixy, k, offset, threshold) 
        cornerList = self.suppressCorners(cornerList, windowsize) 
        finalCorners = self.localizeCorners(cornerList, Ixx, Iyy, Ixy, offset)
        finalCorners.sort()
        
        #Plot detected corner windows
        for i in finalCorners:
            x1 = i[0] - offset
            y1 = i[1] - offset 
            x2 = i[0] + offset + 1
            y2 = i[1] + offset + 1
            plot_img = cv2.rectangle(plot_img, (x1,y1), (x2,y2), [0,0,255], 1)
              
        features = self.featureVectors(finalCorners, dx, dy)
            
        print(len(finalCorners), "corners plotted!\n")
        return plot_img, finalCorners, features
        
    
    #Harris corner detection 
    def harrisCorners(self, height, width, Ixx, Iyy, Ixy, k, offset, threshold):
        cornerList = []
        
        #Scan T-B, L-R
        for y in range(offset, height-offset):
            for x in range(offset, width-offset):
                
                #Get windows
                Wxx = Ixx[y-offset:y+offset+1, x-offset:x+offset+1]
                Wyy = Iyy[y-offset:y+offset+1, x-offset:x+offset+1]
                Wxy = Ixy[y-offset:y+offset+1, x-offset:x+offset+1]
                
                #Calculate sum of squares
                Sxx = Wxx.sum()                
                Syy = Wyy.sum()    
                Sxy = Wxy.sum()                
                
                #Calculate determinant, trace, and cornerness
                det = (Sxx * Syy) - (Sxy**2)
                trace = Sxx + Syy
                g = det - k*(trace**2)
                
                #If cornerness > threshold, add to corner list
                if g > threshold:
                    cornerList.append([g, x, y, 0])               
                    
        return cornerList
    
    
    #Non-maximum suppression
    def suppressCorners(self, cornerList, windowsize):        
        #Sort in decreasing order
        cornerList.sort(reverse=True)
        
        #Mark unwanted neighbors based on Window Size
        for i in cornerList:
            index = []
            if i[3] != 1:
                for j in cornerList:
                    if j[3] != 1:
                        dist = math.sqrt((j[1] - i[1])**2 + (j[2] - i[2])**2)
                        if (dist <= windowsize and dist > 0):
                            j[3] = 1
        
        #Filter out neighbors
        final = filter(lambda x: x[3] == 0, cornerList)   
        
        return list(final)
            
        
    #Corner localization
    def localizeCorners(self, cornerList, Ixx, Iyy, Ixy, offset):
        localizedList = []
        Mgrad = np.zeros((2,2))
        Mcorrelation = np.zeros((2,2))
        
        #go through each corner
        for i in cornerList:
            x = i[1]
            y = i[2]
            Sxx = 0
            Syy = 0
            Sxy = 0
            Evec = np.zeros((2,1))
            point = np.zeros((2,1))
            
            #Scan T-B, L-R
            for row in range(y-offset, y+offset+1):
                for col in range(x-offset, x+offset+1):
                    
                    if (row > len(Ixx)-1 or row > len(Ixy)-1 or row > len(Iyy)-1 
                        or col > len(Ixx[row])-1 or col > len(Ixy[row])-1 or col > len(Iyy[row])-1):
                        continue
                    
                    Mgrad[0,0] = Ixx[row,col]
                    Mgrad[0,1] = Ixy[row,col]
                    Mgrad[1,0] = Ixy[row,col]
                    Mgrad[1,1] = Iyy[row,col]
                    
                    Sxx += Ixx[row,col]
                    Syy += Iyy[row,col]
                    Sxy += Ixy[row,col]
                    
                    point[0,0] = row
                    point[1,0] = col
                    
                    vec = np.dot(Mgrad,point)
                    Evec = np.add(Evec,vec)
                    
            #correlation matrix
            Mcorrelation[0,0] = Sxx
            Mcorrelation[0,1] = Sxy
            Mcorrelation[1,0] = Sxy
            Mcorrelation[1,1] = Syy
            
            #calculate best P
            point = np.dot(np.linalg.pinv(Mcorrelation), Evec).astype('int32')
            localizedList.append([point[1,0],point[0,0]])
            
        return localizedList
        
    
    #Feature extraction
    def featureVectors(self, cornerList, dx, dy):        
        featureList = []
        windowsize = 27 #patch size
        cell_size = 9
        bin_size = 8        
        offset = windowsize//2
        
        #go through each corner
        for i in cornerList:
            feature = []
            x = i[0]
            y = i[1]            
            feature_vector = np.zeros((windowsize//cell_size, windowsize//cell_size, bin_size))
            
            #iterate through each cell in patch
            row=y-offset
            r=0            
            while row < y+offset:                
                col=x-offset
                c=0
                while col < x+offset:
                    #get HOG for each cell
                    feature_vector[r][c] = self.cellHOG(row, col, dx, dy, cell_size, bin_size)
                    col += cell_size
                    c+=1
                row += cell_size
                r+=1
            #add feature for a corner        
            featureList.append(feature_vector)  
            
        return featureList
    
    
    #Get HOG for each cell
    def cellHOG(self, y, x, dx, dy, cell_size, bin_size):
        hog = [0] * bin_size
        
        for row in range(y, y+cell_size):
            for col in range(x, x+cell_size):
                
                if (row > len(dx)-1 or row > len(dy)-1 or col > len(dx[row])-1 or col > len(dy[row])-1):
                    continue

                #get gradient and magnitude
                y_grad = dy[row][col]
                x_grad = dx[row][col]
                grad_mag = np.hypot(x_grad, y_grad)
                if x_grad:
                    angle = math.atan(y_grad / x_grad)
                else:
                    angle = 1.5708 # 90 degrees 
                
                bucket = 6.2831 / bin_size
                i = int(angle / bucket)
                mod = angle % bucket
                if i == bin_size:
                    angle1 = i - 1
                    angle2 = (i) % bin_size
                else:
                    angle1 = i
                    angle2 = (i + 1) % bin_size
                    
                #distribute magnitude
                hog[angle1] += (grad_mag * (1 - (mod / bucket)))
                hog[angle2] += (grad_mag * (mod / bucket))
                
        #rotate to start with dominant direction
        m = hog.index(max(hog))
        hog = hog[m:] + hog[:m] 
        
        return hog
    
    
    #Feature comparison
    def compareFeatures(self, plot_img1, finalCorners1, features1, plot_img2, finalCorners2, features2):   
        if (len(features1) == 0 or len(features2) == 0):
            return
        map_features=[]
        
        #match corners based on closeness of feature vectors
        for i in range(0,len(finalCorners1)):
            c1 = finalCorners1[i]
            f1 = features1[i]
            all_dist_compare = []
            
            for j in range(0,len(finalCorners2)):
                c2 = finalCorners2[j]
                f2 = features2[j]  
                total_dist=0
                
                for u in range(0,len(f1)):
                    for v in range(0,len(f1[u])):
                        dist = np.linalg.norm(f1[u][v]-f2[u][v])
                        total_dist += dist
                all_dist_compare.append(total_dist)
                
            mindist = min(all_dist_compare)
            m = all_dist_compare.index(mindist)
            map_features.append(m)
            
        #mapping list
        final_map = []
        for i in range(0,len(finalCorners2)):            
            final_map.append([(finalCorners2[i][0],finalCorners2[i][1]),[]])
            
        for i in range(0,len(finalCorners1)):
            final_map[map_features[i]][1].append((finalCorners1[i][0],finalCorners1[i][1]))
            
        #plot numbers
        counter = 1
        font = cv2.FONT_HERSHEY_SIMPLEX 
        color = (55, 255, 55) 
        fontScale = 0.7 
        thickness = 1
        for i in range(0,len(final_map)):
            if len(final_map[i][1]) == 0:
                continue
            
            for p in final_map[i][1]:
                org = p 
                plot_img1 = cv2.putText(plot_img1, str(counter), org, font,  
                                   fontScale, color, thickness, cv2.LINE_AA)
                
            org = (final_map[i][0]) 
            plot_img2 = cv2.putText(plot_img2, str(counter), org, font,  
                               fontScale, color, thickness, cv2.LINE_AA) 
            counter+=1
        
        print("Number of matched features: ", counter-1, '\n')
        self.curr_img = plot_img1
        self.curr_img2 = plot_img2
        cv2.imshow(self.TITLE, self.curr_img)
        cv2.imshow(self.TITLE2, self.curr_img2)
```
